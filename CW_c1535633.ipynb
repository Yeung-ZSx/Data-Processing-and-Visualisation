```python
###cell 3###=======================================================================================================

# read in image data
# functions to read in and display an image
query_img_size = figure(figsize(2,1))
imshow(img_query)
title('query')
axis('off') 

red_Hist = numpy.histogram(r, bins = arange(0,256,1), normed = True)
green_Hist = numpy.histogram(g, bins = arange(0,256,1), normed = True)
blue_Hist = numpy.histogram(b, bins = arange(0,256,1), normed = True)
    
Euclidean_List = []
j=0
# %int(i) for i in range(1,31)
for img_database in ('./IRD/%d.jpg'%int(i) for i in range(1,31)):
    
    img=imread(img_database)
    
    Red_database_img = img[:,:,0]
    Green_database_img = img[:,:,1]
    Blue_database_img = img[:,:,2]
    
    Red_db = Red_database_img.ravel()      
    Green_db = Green_database_img.ravel()       
    Blue_db = Blue_database_img.ravel()

    Red_db_Hist = numpy.histogram(Red_db, bins = arange(0,256,1), normed = True)
    Green_db_Hist = numpy.histogram(Green_db, bins = arange(0,256,1), normed = True)
    Blue_db_Hist = numpy.histogram(Blue_db, bins = arange(0,256,1), normed = True)
    
    Euclidean_Red = sum([(a-b)**2 for a, b in zip(Red_db_Hist[0], red_Hist[0])])
    Euclidean_Green = sum([(a-b)**2 for a, b in zip(Green_db_Hist[0], green_Hist[0])])
    Euclidean_Blue = sum([(a-b)**2 for a, b in zip(Blue_db_Hist[0], blue_Hist[0])])
    Dist_of_Euclidean = round(sqrt(Euclidean_Red+Euclidean_Green+Euclidean_Blue),4)
    Euclidean_List.append((img,Dist_of_Euclidean))

eu_fig=figure(figsize(20,5))
Euclidean_List.sort(key=lambda j: j[1])
for i, j in enumerate(Euclidean_List):
    subplot(3,10,i+1)
    imshow(j[0]), axis('off')
    title("s=%.4f" % j[1])
    

###cell 3###=======================================================================================================
```
