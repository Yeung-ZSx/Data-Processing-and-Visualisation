```python
%pylab inline
import pandas as pd
from pandas import Series, DataFrame
from scipy import stats
```

```python
# read in from an Microsoft Excel file
xls_file=pd.ExcelFile('LN06_handout.xlsx')
data=xls_file.parse('Sheet1')

# Means and standard deviations
cw=data['CW']
ex=data['EXM']
cw_mean=cw.mean()
cw_std=cw.std()
ex_mean=ex.mean()
ex_std=ex.std()
print('coursework average mark:', round(cw.mean(),1))
print('coursework standard deviation:', round(cw.std(),1))
print('exam average mark:', round(ex.mean(), 1))
print('exam standard deviation:', round(ex.std(), 1))
```

```python
# bar plot
figure(figsize(12,6))
N=2
ind = arange(N)  # the x locations for the data groups
width = 0.35    # the width of the bars
bar(ind, [cw_mean, ex_mean], width, color='r', yerr=[cw_std, ex_std])
ylabel('marks', fontsize=30)
xtickloc = ind+(width/2)
xticklabels = ['CW', 'EXM']
xticks(xtickloc, xticklabels, rotation='horizontal', fontsize=20)
title('CW&EXAM marks', fontsize=30)
xlim([-0.5, 2])
ylim([0, 100])
plt.show()
```

```python
# Write code to process the USA presidential campaign contributions dataset
# 1) read in the data file named 'P-UT.csv'
# 2) select four candidates’ donations: 'Obama, Barack', ‘McCain, John S’, 'Paul, Ron' and 'Clinton, Hillary Rodham'
# 3) visualise the boxplots of candidates’ donations
# 4) visualise the bar plots of candidates’ donations

# Your code==================================================================================================









# Your code==================================================================================================
```

```python
# Solutions==================================================================================================
# read in data file
data=pd.read_csv('P-UT.csv', index_col=False)

# extract required data
ob=data[data['cand_nm']=='Obama, Barack']
ob=ob['contb_receipt_amt']
ob=array(ob)
mc=data[data['cand_nm']=='McCain, John S']
mc=mc['contb_receipt_amt']
mc=array(mc)
pa=data[data['cand_nm']=='Paul, Ron']
pa=pa['contb_receipt_amt']
pa=array(pa)
cl=data[data['cand_nm']=='Clinton, Hillary Rodham']
cl=cl['contb_receipt_amt']
cl=array(cl)

# 
N=4
ind = arange(N)  # the x locations for the four categories 

# Boxplot
figure(figsize(10,5))
boxplot([ob, mc, pa, cl])
xtickloc = ind+1
xticklabels = ['Obama', 'McCain', 'Paul', 'Clinton']
xticks(xtickloc, xticklabels, rotation='horizontal', fontsize=20)
ylabel('donations', fontsize=20)
title('Candidates donations', fontsize=20)

# Bar plot
ob_mean=ob.mean()
ob_std=ob.std()
mc_mean=mc.mean()
mc_std=mc.std()
pa_mean=pa.mean()
pa_std=pa.std()
cl_mean=cl.mean()
cl_std=cl.std()

figure(figsize(10,5))
width = 0.35       # the width of the bars
bar(ind,[ob_mean, mc_mean, pa_mean, cl_mean], width, color='r', yerr=[ob_std, mc_std, pa_std, cl_std])
xtickloc = ind+(width/2)
xticklabels = ['Obama', 'McCain', 'Paul', 'Clinton']
xticks(xtickloc, xticklabels, rotation='horizontal', fontsize=20)
ylabel('average donation', fontsize=20)
title('Candidates donations', fontsize=20)
# Solutions==================================================================================================
```

```python
# Solutions==================================================================================================
# Bar plot - grouping
figure(figsize(10,5))
width = 0.35       # the width of the bars
gp=arange(N/2)     # number of groups
plot1=bar(gp,[ob_mean, mc_mean], width, color='y', yerr=[ob_std, mc_std])
plot2=bar(gp+width,[pa_mean, cl_mean], width, color='r', yerr=[pa_std, cl_std])

xtickloc = gp+width
xticklabels = ['Obama vs Paul', 'McCain vs Clinton']
xticks(xtickloc, xticklabels, rotation='horizontal', fontsize=20)
title('Candidates donations', fontsize=20)
ylabel('average donation', fontsize=20)

legend((plot1[0], plot2[0]), ('Party1', 'Party2'))

plt.show()
# Solutions==================================================================================================
```
